Обязательная часть

Будем парсить страницу со свежеми новостям на habr.com/ru/all/.

Вам необходимо собирать только те статьи, в которых встречается хотя бы одно требуемое ключевое слово. Эти слова определяем в начале кода в переменной, например:

KEYWORDS = ['python', 'парсинг']

Поиск вести по всей доступной preview-информации (это информация, доступная непосредственно с текущей страницы).

В итоге должен формироваться датафрейм вида: <дата> - <заголовок> - <ссылка>


```python
import requests
```

```python
import time
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
URL = 'https://habr.com/ru/search/'
params = {
    'q': ['python', 'парсинг'],
}
```

```python
res = requests.get(URL, params)
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
news_blocks = soup.find_all('div', class_='page__body')
```

```python
#даты
articles_date = list(map(lambda x: x.find_all('span', class_='post__time'), news_blocks))
```

```python
#заголовки
articles_heading = list(map(lambda x: x.find_all('h2', class_='post__title'), news_blocks))
```

```python
#ссылки

all_links = []

for i in articles_heading:
    for x in i:
        y = x.find('a').get('href')
        all_links.append(y)
        
        
print(all_links)
```

```python
# запишем данные в датафрейм
habr_news = pd.DataFrame()

for link in all_links:
    soup = BeautifulSoup(requests.get(link).text, 'html.parser')
    time.sleep(0.3)
    
    date = soup.find('span', class_='post__time').get('data-time_published')
   
    title = soup.find('span', class_='post__title-text').text
    
    row = {'date': date, 'title': title, 'link': link}
    
    habr_news = pd.concat([habr_news, pd.DataFrame([row])]) 
    
habr_news
```

# Задание 2

Написать скрипт, который будет проверять список e-mail адресов на утечку при помощи сервиса Avast Hack Ckeck. Список email-ов задаем переменной в начале кода:
EMAIL = [xxx@x.ru, yyy@y.com]

В итоге должен формироваться датафрейм со столбцами: <почта> - <дата утечки> - <источник утечки> - <описание утечки>

Подсказка: сервис работает при помощи "скрытого" API. Внимательно изучите post-запросы.


```python
import json
```

```python
email = ['xxx@x.ru', 'yyy@y.com']

url = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'

params = {'emailAddresses': email}

headers = {'Vaar-Version' : '0', 'Vaar-Header-App-Product' : 'hackcheck-web-avast'}

req = requests.post(url, data = json.dumps(params), headers = headers)
```

```python
dict_ = req.json()
```

```python
df = pd.DataFrame()
```

```python
for k,v in dict_.items():
      if k == 'summary':
        for p, d in dict_['summary'].items(): 
            email_ = p
            list_ = d['breaches']
            
            for i in list_:
                date = (dict_['breaches'][str(i)]['publishDate'])
    
                source = (dict_['breaches'][str(i)]['site'])
    
                description = (dict_['breaches'][str(i)]['description'])
    
                row = {'почта': email_, 'дата утечки': date, 'источник утечки': source, 'описание утечки': description}
    
                df = pd.concat([df, pd.DataFrame([row])])
```
