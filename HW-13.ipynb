# Задание 1
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
r1 = ratings.groupby('userId').agg({'rating': 'count', 'timestamp':['min', 'max']})
```

```python
r1['lifetime'] = r1['timestamp', 'max'] - r1['timestamp', 'min']
```

```python
r2 = r1.loc[r1.rating ['count'] > 100]

# r2 — датафрейм, в котором только пользователи поставившиее больше 100 оценок. 
# Для каждого пользователя посчитано время жизни.
```

```python
mean_lifetime = r2['lifetime'].mean()

print(f'Cреднее время жизни пользователей, которые выставили более 100 оценок = {mean_lifetime}')
```
